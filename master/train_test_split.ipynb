# Train/Dev/Test Split

```
# Get a list of unique question IDs
unique_question_ids = np.array(df['question_id'].unique())
num_unique_questions = len(unique_question_ids)
print("length:",num_unique_questions)
unique_question_ids
```

```

# Calculate number of questions for splits
num_questions_80_percent = math.floor((80 * num_unique_questions) / 100)
num_questions_10_percent = int((num_unique_questions - num_questions_80_percent) / 2)
print("80%%: %d    10%%: %0d  "%(num_questions_80_percent,num_questions_10_percent))
```

```
# Perform 80/10/10 split
random.seed(1)
train_indices = random.sample(range(0, num_unique_questions), num_questions_80_percent)
train_question_ids = list(unique_question_ids[train_indices])
questions_remaining_after_train = np.delete(unique_question_ids, train_indices)
dev_indices = random.sample(range(0, len(questions_remaining_after_train)), num_questions_10_percent)
dev_question_ids = list(questions_remaining_after_train[dev_indices])
test_question_ids = list(np.delete(questions_remaining_after_train, dev_indices))
print("train length:", len(train_question_ids), train_question_ids)
print("dev length:   ", len(dev_question_ids), dev_question_ids)
print("test length:  ", len(test_question_ids), test_question_ids)
```

```
# Get train, dev, and test dataframes
train_data = df[df['question_id'].isin(train_question_ids)]
dev_data = df[df['question_id'].isin(dev_question_ids)]
test_data = df[df['question_id'].isin(test_question_ids)]
print("Train Data: ", train_data.shape)
print("Dev Data: ", dev_data.shape)
print("Test Data: ", test_data.shape)
```
