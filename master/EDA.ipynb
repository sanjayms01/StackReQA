## Token Statistics

```
from collections import defaultdict

#Creating dictionaries of tokens:count
question_title_tokens = defaultdict(lambda: 0.0)
question_text_tokens = defaultdict(lambda: 0.0)
question_code_tokens = defaultdict(lambda: 0.0)
answer_text_tokens = defaultdict(lambda: 0.0)
answer_code_tokens = defaultdict(lambda: 0.0)

for i in range(1,df.shape[0]):

  for token in df.question_title_raw[i].split(" "):
    question_title_tokens[token] += 1
  
  for token in str(df.question_body_text[i]).split(" "):
      question_text_tokens[token] += 1

  for token in str(df.question_body_code[i]).split(" "):
      question_code_tokens[token] += 1
  
  for token in str(df.answer_text[i]).split(" "):
      answer_text_tokens[token] += 1
  
  for token in str(df.answer_code[i]).split(" "):
      answer_code_tokens[token] += 1

#Combining tokens from all 5 columns above
combined_tokens=defaultdict(lambda:0.0)
for key,value in question_title_tokens.items():
  combined_tokens[key] += value
for key,value in question_text_tokens.items():
  combined_tokens[key] += value
for key,value in question_code_tokens.items():
  combined_tokens[key] += value
for key,value in answer_text_tokens.items():
  combined_tokens[key] += value
for key,value in answer_code_tokens.items():
  combined_tokens[key] += value
```

```
#Sorting Tokens by k Most Frequent

k=10
question_title_tokens_topk = sorted(question_title_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
question_text_tokens_topk = sorted(question_text_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
question_code_tokens_topk = sorted(question_code_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
answer_text_tokens_topk = sorted(answer_text_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
answer_code_tokens_topk = sorted(answer_code_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
combined_tokens_topk = sorted(combined_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]

print("Question Title Tokens: %d" %len(question_title_tokens))
print("Top %d:" %(k), question_title_tokens_topk)
print("")
print("Question Text Tokens: %d" %len(question_text_tokens))
print("Top %d:" %(k), question_text_tokens_topk)
print("")
print("Question Code Tokens: %d" %len(question_code_tokens))
print("Top %d:" %(k), question_code_tokens_topk)
print("")
print("Answer Text Tokens: %d" %len(answer_text_tokens))
print("Top %d:" %(k), answer_text_tokens_topk)
print("")
print("Answer Code Tokens: %d" %len(answer_code_tokens))
print("Top %d:" %(k), answer_code_tokens_topk)
print("")
print("Combined Tokens: %d" %len(combined_tokens))
print("Top %d:" %(k),combined_tokens_topk )
```

```
# K most popular Code Tokens

k=50
question_code_tokens_topk = sorted(question_code_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]
answer_code_tokens_topk = sorted(answer_code_tokens.items(), key = lambda kv: kv[1], reverse=True)[0:k]

print("Question Code Tokens: %d" %len(question_code_tokens))
print("Top %d:" %(k), question_code_tokens_topk)
print([x[0] for x in question_code_tokens_topk])
print("")
print("Answer Code Tokens: %d" %len(answer_code_tokens))
print("Top %d:" %(k), answer_code_tokens_topk)
print([x[0] for x in answer_code_tokens_topk])
```

# Checking Question/Answer Body Outlinks

```
df_links = pd.read_csv (r'/content/drive/MyDrive/Colab Notebooks/W266 Final Project/data/raw_data_preprocessed.csv')
df_links
```

```
from bs4 import BeautifulSoup

row_indices = []

for i in df_links.index:
  body = df_links.iloc[i]['question_body_raw']
  soup = BeautifulSoup(body, 'html.parser')
  if len(soup.find_all('a')) >= 1:
    row_indices.append(i)

print(len(row_indices))
```
